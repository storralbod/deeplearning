```python
import os
import requests
from datetime import datetime, timedelta

# Base URL and directory to save the files
base_url = "https://www.omie.es/en/file-download?parents%5B0%5D=precios_pibcic&filename=precios_pibcic_{}.1"
output_dir = "./data_24"

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Function to download a single file
def download_file(date_str):
    url = base_url.format(date_str)
    file_name = f"precios_pibcic_{date_str}.1"
    output_path = os.path.join(output_dir, file_name)
    
    # Skip if file already exists
    if os.path.exists(output_path):
        print(f"File already exists: {file_name}")
        return
    
    try:
        response = requests.get(url, timeout=30)
        response.raise_for_status()  # Raise an error for HTTP issues
        with open(output_path, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded: {file_name}")
    except requests.exceptions.RequestException as e:
        print(f"Failed to download {file_name}: {e}")

# Loop over dates and download files
start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 11, 30)
current_date = start_date

while current_date <= end_date:
    date_str = current_date.strftime("%Y%m%d")
    download_file(date_str)
    current_date += timedelta(days=1)

print("All downloads completed.")
```
